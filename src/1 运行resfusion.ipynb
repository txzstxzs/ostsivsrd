```python
def resfusion_x0_to_xt(x_0: torch.Tensor, alpha_hat_t: torch.Tensor, residual_term: torch.Tensor, noise: torch.Tensor) \
        -> torch.Tensor:
    """
    Compute x_t from x_0 with formula 20 from Resfusion
    :param x_0: the image without noise
    :param alpha_hat_t: the cumulated variance schedule at time t
    :param residual_term: the residual term
    :param noise: pure noise from N(0, 1)
    :return: the noised image x_t at step t
    """
    return (torch.sqrt(alpha_hat_t) * x_0 + (1 - torch.sqrt(alpha_hat_t)) * residual_term
            + torch.sqrt(1 - alpha_hat_t) * noise)


def ddpm_x0_to_xt(x_0: torch.Tensor, alpha_hat_t: torch.Tensor, noise: torch.Tensor) -> torch.Tensor:
    """
    Compute x_t from x_0 using a closed form using theorem from original DDPM paper (Ho et al.)
    :param x_0: the image without noise
    :param alpha_hat_t: the cumulated variance schedule at time t
    :param noise: pure noise from N(0, 1)
    :return: the noised image x_t at step t
    """
    return torch.sqrt(alpha_hat_t) * x_0 + torch.sqrt(1 - alpha_hat_t) * noise
```

```python
epoch = 200
n_iters = epoch * ( len(training_data) )

iters = 0
mr = 0.3

loss_list=[]
net = SSSDS4Imputer(**model_config).cuda()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)
```

```python
'掩码产生过程和训练'
'-----这里每个batch里 每个数据的掩码都一样 不同batch里数据的掩码不一样-----'
'不同的缺失率要另外训练'
'无监督任务的话缺失掩码应该没有用'

while iters < n_iters + 1:         
    for batch in training_data:           # [128, 30, 6]
        
        X_0 = batch.float().cuda()  
        B = X_0.shape[0]
        L = X_0.shape[1]
        D = X_0.shape[2]
        
        '退化数据的几种选择 全0 噪声 缺失数据'
        '如果用缺失数作为退化数据 缺失部分可以看成0 那么计算resnoise损失是就不用加掩码'
        '也可以计算resnoise损失时加上掩码 只计算观测部分的损失'
#         X_0_hat = torch.zeros_like(X_0)   #  全0 [128, 3, 32, 32]  可以选择全0或噪声
#         X_0_hat = torch.randn_like(X_0)   #  全1

        '4种缺失场景'
        missing_length = int( mr * L )
#         mask = get_mask_rm( X_0[0], missing_length)      # 原数据乘以缺失掩码
#         mask = get_mask_mnr( X_0[0], missing_length) 
#         mask = get_mask_bm( X_0[0], missing_length) 
        mask = get_mask_mcar( X_0[0], mr)  # mcar缺失法的输入参数和另外单个不一样 直接输入缺失率
    
        mask = mask.repeat(X_0.size()[0], 1, 1).float().cuda()
        X_0_hat = X_0*mask

        residual_term = X_0_hat - X_0 
        
        t: torch.Tensor = torch.randint(0, T_acc, (X_0.shape[0],) ).cuda()     #  
        alpha_hat = Alpha_bar[t].reshape(-1, 1, 1).cuda()                      # 
        noise = torch.randn_like(X_0)
        
        x_t = resfusion_x0_to_xt(X_0, alpha_hat, residual_term, noise) 
        
        alpha = Alpha[t].reshape(-1, 1, 1).cuda() 
        beta = Beta[t].reshape(-1, 1, 1).cuda() 
        
        resnoise = noise + (1 - torch.sqrt(alpha)) * torch.sqrt(1 - alpha_hat) / beta * residual_term
        
    
        '两种训练模式 有条件和无条件 有条件是融入观测值和掩码 二者会拼接'
        pred_resnoise = net( (x_t.permute(0, 2, 1), X_0.permute(0, 2, 1), mask.permute(0, 2, 1), t.view(B, 1),) )  #  [B, 6, 30]
#         pred_resnoise = net( (x_t.permute(0, 2, 1), None, None, t.view(B, 1),) ) 
        pred_resnoise = pred_resnoise.permute(0, 2, 1)
        
        optimizer.zero_grad()
        
        '训练时用mse作为损失好像性能好一些'
        '或许损失应该只计算观测位置'

        loss = F.mse_loss(input=pred_resnoise, target=resnoise )  
#         loss = F.mse_loss(input=pred_resnoise*(1-mask), target=resnoise*(1-mask) )  
#         loss = torch.sqrt( F.mse_loss(input=pred_resnoise, target=resnoise) )  # 损失可以除以batchsize
#         loss = F.smooth_l1_loss(input=pred_resnoise, target=resnoise)
    
        loss.backward()
        optimizer.step()
        loss_list.append(loss.item())
        iters += 1
         
```

```python
'----------reflow-----------'

import torch
from torch.nn import functional as F
from torch.distributions.normal import Normal


def flow_matching_loss(student_model, x_0, x_1, t, encoder_hidden_states):
    t = t.reshape((t.shape[0], 1, 1, 1))
    x_t = t * x_1 + (1 - t) * x_0
    t = t.flatten()
    
    student_pred = student_model(x_t, t, encoder_hidden_states=encoder_hidden_states).sample
    
    return F.mse_loss(student_pred, (x_1 - x_0), reduction="mean")


'欧拉采样过程 需要搞清楚x0和x1分别是什么'
'按照下面的流程的话x0应该是退化图像'
def sample_x1(unet, vae, x_0, encoder_hidden_states, n_steps: int = 100, device: str = "cuda"):
    unet.to(device)
    vae.to(device)
    unet.eval()
    vae.eval()
    
    x_t = x_0.to(device)
    bs = x_0.shape[0]
    eps = 1e-8         # t从一个很接近0的值开始
    t = torch.linspace(eps, 1 - eps, n_steps + 1).to(device)

    for i in range(1, len(t)):
        t_prev = t[i - 1].unsqueeze(0).repeat((bs,))

        with torch.no_grad():          # 计算速度
            f_eval = unet(x_t, t_prev, encoder_hidden_states=encoder_hidden_states).sample
            
        x_t = x_t + (t[i] - t[i - 1]) * f_eval   # xt加上时间间隔 dt乘以预测的速度
    
    return x_t


# t = torch.sigmoid(torch.randn((bs,), device=device))         # 随机的t
# torch.nn.utils.clip_grad_norm_(student_unet.parameters(), 1.0) 
```

```python
from scipy.integrate import quad

# 计算两个值之间的速度  用两个值之差 除以时间间隔 得到速度  
def compute_velocity(z_t, z_t_minus_1, delta_t):
    v_t = (z_t_minus_1 - z_t) / delta_t
    return v_t
                         # (248, 1, 4, 32, 32)  上面的函数     false
def straightness_measurement(Z_t_list, compute_velocity, use_velocity_fn):
    """
    Measure the straightness of a trajectory given initial and final positions (Z_0, Z_1)
    and a velocity function v(Z_t, t).
    """

#     Z_0 = Z_t_list[0]   # 估计的起点和终点
#     Z_1 = Z_t_list[-1]
    
    
    Z_0 = Z_t_list[-1]   # 修改方向 这个要改的话  上面的compute_velocity里的两个相减的值就不用换位置
    Z_1 = Z_t_list[0]
    

    def integrand(t):
        # t is [0, 1], we must scale it to the length of Z_t_list
        i = t * (len(Z_t_list) - 1)
        delta_t = 1 / len(Z_t_list) # 时间间隔 for now we assume linearly spaced time steps

        if use_velocity_fn:            # 当前为false 不用额外计算向量
            if t == 0:
                v_t = 0
            else:                  # 计算相邻两点直接的速度
                v_t = compute_velocity(Z_t_list[int(i)], Z_t_list[int(i-1)], delta_t)  
        else:
            v_t = Z_t_list[int(i)]     # 直接取向量值
            
        deviation = (Z_1 - Z_0) - v_t
        deviation = np.array(deviation.cpu())  #要转换成np格式才好进行下面的取均值计算
        
        return np.mean(deviation ** 2)

    # Integrate from t=0 to t=1
    S_Z, _ = np.log(quad(integrand, 0, 1))
    return S_Z


straightness_value_ddpm = straightness_measurement( trajectory_ddpm, compute_velocity, use_velocity_fn=True)
straightness_value_ddim = straightness_measurement( trajectory_ddim, compute_velocity, use_velocity_fn=True)
straightness_value_reflow = straightness_measurement( trajectory_reflow, compute_velocity, use_velocity_fn=True)
print(straightness_value_ddpm, straightness_value_ddim, straightness_value_reflow )
```
